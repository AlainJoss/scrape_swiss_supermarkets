```python
# sort by price_per_unit and display top 10

attribute = 'price_per_unit'
attribute = 'protein (g)'

subset = ['name', 'url', 'price', 'price_per_unit', 'protein (g)']

df.sort_values(by=attribute, ascending=False)[subset].head(20)
```

```python
import matplotlib.pyplot as plt

# Filter out items with price_per_unit > 500 (remove outliers)
filtered_df = df[df['price_per_unit'] <= 50]

# Plot price_per_unit against protein (g) after filtering outliers
filtered_df.plot.scatter(x='protein (g)', y='price_per_unit', title='Price per Unit vs Protein (g) (Outliers Removed)', alpha=0.7)
plt.xlabel('Protein (g)')
plt.ylabel('Price per Unit')
plt.show()

# Calculate the ratio of price_per_unit to protein (g)
df['price_to_protein_ratio'] = df['protein (g)'] / df['price_per_unit'] 

# Sort by the ratio and display the top 10 items
top_ratio = df.sort_values(by='price_to_protein_ratio', ascending=False)[subset]
# threshold protein values over 10
top_ratio = top_ratio[top_ratio['protein (g)'] > 20]
print(top_ratio.head(10))
```
